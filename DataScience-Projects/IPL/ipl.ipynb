```python
# Getting the frequency of most man-of-match awards
ipl['Player_of_Match'].value_counts()
```

```python
# Getting the top 10 players with most man-of-match awards
ipl['Player_of_Match'].value_counts()[0:10]
```

```python
# Getting the top 5 players with most of the match awards
man_of_match = ipl['Player_of_Match'].value_counts()
man_of_match=man_of_match.head()
man_of_match
```

```python
# Make Bar-Plot for the top 5 players with most man-of-match awards
man_of_match=dict(man_of_match)
x = list(man_of_match.keys())
y = list(man_of_match.values())
plt.bar(x,y,color=['g','r','b','y','m'])
plt.show()
# x,y
```

```python
# Getting the frequency of result column
ipl['WinningTeam'].value_counts()
```

```python
# Finding out the number of toss wins w.r.t each team
toss_winner = ipl['TossWinner'].value_counts()
toss_winner
```

```python
# Extracting the records where a team won toss and choose to bat first
batting_toss_winner = ipl[(ipl['TossDecision'] == 'bat')]
batting_toss_winner = batting_toss_winner['TossWinner']
display(batting_toss_winner.value_counts())
print("Total Teams who won the toss and bat first: ",batting_toss_winner.shape[0])
```

```python
# Extracting the records where a team won toss and choose to bat first and also win the match also
# batting_first = ipl[(ipl['TossDecision'] == 'bat')]
# batting_toss_game_winner = batting_first[(batting_first['WinningTeam'] == batting_first['TossWinner'])]
# Same result can be achieved by:
batting_toss_game_winner = ipl[(ipl['TossDecision'] == 'bat') & (ipl['TossWinner'] == ipl['WinningTeam'])]

Winner_teams = batting_toss_game_winner['WinningTeam']
display(Winner_teams.value_counts())
print('Total GAMES who won toss and bat first and also game are: ',Winner_teams.shape[0])
```

```python
# Extracting the records where a team won toss and choose to field first and also win the match also
bowling_toss_game_winner = ipl[(ipl['TossDecision'] == 'field') & (ipl['TossWinner'] == ipl['WinningTeam'])]
bowling_toss_game_winner.shape
```

```python
# Making a Bar Chart
# plt.bar(['Toss_Bat_Game_Winner','Total Games'],[batting_toss_game_winner.shape[0],ipl.shape[0]])
plt.bar(['Batting_First_Winners','Bowling_First_Winners'],[batting_toss_game_winner.shape[0],bowling_toss_game_winner.shape[0]])
plt.show()
```

```python
# Making a pie chart
winners = ['Batting_First_Winners','Bowling_First_Winners']
values = [batting_toss_game_winner.shape[0],bowling_toss_game_winner.shape[0]]
plt.pie(values,labels=winners)
plt.show()
```

```python
# Making a bar-plot for top 3 teams with most wins after batting first
total_winning_games = batting_toss_game_winner['WinningTeam'].value_counts()
teams = total_winning_games.head(3)
# teams.columns
x=teams.index.tolist()
y=teams.values.tolist()
plt.bar(x,y)
plt.show()
```

```python
# Making a bar-plot for top 3 teams with most wins after `bowling-first`
top_winners=bowling_toss_game_winner['WinningTeam'].value_counts().head(3)
x=top_winners.index.tolist()
y=top_winners.values.tolist()
plt.bar(x,y)
plt.show()
```

### First Batting Winners

```python
# Finding out the number of wins w.r.t each team after batting first except toss
bat_first_winners = ipl[ipl['WonBy']=='Runs']
bat_first_winners['WinningTeam'].value_counts()
```

```python
# Show distribution of winning by how many-runs by histogram
# plt.figure(figsize=(5,5))
plt.hist(bat_first_winners['Margin'],bins=150)
plt.show()
bat_first_winners['Margin'].value_counts()
```

```python
# Making a bar-plot for top 3 teams with most wins after `batting-first`
x = bat_first_winners['WinningTeam'].value_counts()[0:3].index.tolist()
y = bat_first_winners['WinningTeam'].value_counts()[0:3].values.tolist()
plt.bar(x,y,color=['r','y','b'])
plt.show()
```

```python
# Making a pie-chart
labels = bat_first_winners['WinningTeam'].value_counts().index.tolist()
values = bat_first_winners['WinningTeam'].value_counts().values.tolist()
plt.pie(values,labels=labels,autopct="%0.1f%%") # autopct->For distribution of percentage
plt.show()
```

### Bowling Won

```python
# Show teams who won when bowling first
bowling_first_winners = ipl[ipl['WonBy']=='Wickets']
bowling_first_winners['WinningTeam'].value_counts()
```

```python
# Show distribution of how many-wickets win by histogram
plt.figure(figsize=(7,7))
plt.hist(bowling_first_winners['Margin'],bins=30)
plt.show()
bowling_first_winners['Margin'].value_counts()
```

```python
# Making a bar-plot for top 3 teams with most wins after `batting-first`
x = bowling_first_winners['WinningTeam'].value_counts()[0:3].index.tolist()
y = bowling_first_winners['WinningTeam'].value_counts()[0:3].values.tolist()
plt.bar(x,y,color=['r','y','g'])
plt.show()
```

```python
# Making a pie-chart to show distribution of teams won most by bowling first

labels = bowling_first_winners['WinningTeam'].value_counts().index.tolist()
values = bowling_first_winners['WinningTeam'].value_counts().values.tolist()

# labels,values
plt.pie(values,labels=labels,autopct="%1.1f%%")
plt.show()
```
