#### Apply Custom Function On Columns(Series Data)

```python
def grade(num:int)->str:
    if num >= 80:
        return "Super A+"
    elif num >= 70:
        return "A"
    elif num >= 60:
        return "B"
    elif num >= 50:
        return "C"
    elif num >= 40:
        return "D"
    elif num >= 30:
        return "E"
    else:
        return "Fail"
    
df['Grade'] = df['Percentage'].apply(grade)
print(df.head())
print("============================================================")
print(df.tail())
```

#### Apply Custom Functions on Multiple Columns and return Single or Multiple Columns

* iloc

* iat

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df1 = pd.DataFrame(data, index=['row1', 'row2', 'row3'])

# Accessing a single value using at()
display(df1)
value = df1.at['row2', 'B']
print("======================================")
print("Value at row 'row2' and column 'B':", value)
print("======================================")
df1.loc["row2":"row3","A":"B"]
```

```python
def obtained_marks(s1:int,s2:int,s3:int,s4:int,s5:int)->int:
    return s1+s2+s3+s4+s5
# def obtained_marks(s1:int)->int:
    # return s1

df['Obtained_Marks'] = df[['s1','s2','s3','s4','s5']].apply(lambda x : obtained_marks(*x),axis=1)
df
```

```python
def fn_obtained(s1:int,s2:int,s3:int,s4:int,s5)->tuple[int,int,float,str]:
    total = 500
    obtained = s1+s2+s3+s4+s5
    percentage = (obtained/total)*100
    grade = ""

    if percentage >= 80:
        grade = "A+"
    elif percentage >= 70:
        grade = "A"
    elif percentage >= 60:
        grade = "B"
    elif percentage >= 50:
        grade = "C"
    elif percentage >= 40:
        grade = "D"
    elif percentage >= 30:
        grade = "E"
    else:
        grade = "F"

    return obtained,total,percentage,grade


df[['Obtained','Total','Percentage','Grade']] = df[['s1','s2','s3','s4','s5']].apply(lambda x : fn_obtained(*x),axis=1,result_type='expand') 
df   
```

#### Map Function for Remarks

```python
from typing import Dict

remark : Dict[str,str] = {
    "A+": "Excellent",
    "A": "Great",
    "B": "Good",
    "C": "Keep it up",
    "D": "Average",
    "E": "Just Pass",
    "F": "Fail"
}


df['Remark'] = df['Grade'].map(remark)
df
```
