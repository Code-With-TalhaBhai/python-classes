## Error Handling

* try
    * Logic
* except(error_class1,error_class2)
    * If error occured then run this block
* else:
    * If error not occured
* finally:
    * Always runs

```python
# logic after error not working as Python runs on interpreter
print("logic before error")
7/0
print("logic after error")
```

```python
try:
    print(7/0)   
except(ZeroDivisionError):
    print("You got an error. Please check your input")  
else:
    print("Else block is working")    
finally:
    print("Finally block always running")                
```

```python
try:
    print(0/7)
except(ZeroDivisionError):
    print("Except working error")
else:
    print("Else block running. NO error")
finally:
    print('Finally block always working')
```

```python
l1 = [1,2,3,4,5]
try:
    l1[100]
except ZeroDivisionError:
    print("Its a zero division error")
except IndexError:
    print("Its an index error running from second except")
finally:
    print("Always running")
```

```python
try:
    xyz
except ZeroDivisionError:
    print("Its a zero division error")
except NameError:
    print("Its a name error. Please enter correct variable")
finally:
    print("Always running")
```

```python
l1 = [234,432,52,35,25,235]

try:
    print(l1[342])
except Exception as e:
    print(f"Please correct this error.{e}")
```

```python
class StudentError():
    def __init__(self,roll_no:int,name:str,age:int):
        if age < 18 or age > 23:
            raise StudentAgeError("You are not eligible for this post. You have age issues")
        self.age = age
        self.roll_no = roll_no
        self.name = name


class StudentAgeError(Exception):
    pass


student1 = StudentError(234,'Talha',20)
print(student1.name,student1.roll_no,student1.age)
```

```python
with open('./first.txt') as file: # `with open` closes file automatically
    print(file.read())
```

```python
with open('./first.txt') as file:
    print(file.readline()) # Read line-by-line
    print(file.readline())
```

```python
with open('first.txt') as file:
    print(file.readlines())
```

```python
# Read characters from index
with open('./first.txt') as file:
    print(file.readline()[3:])
```

```python
# Read lines from index
with open('./first.txt','r') as file:
    print(file.readlines()[1:])
```

```python
with open('first.txt','w+') as file:
    file.write('I Love Pakistan')
```

```python
with open('first.txt','r+') as file:
    print(file.read())
    file.write(' I Again love Pakistan')
    print(file.read())
```

```python
with open('./first.txt','r+') as file:
    print(file.read())
    
    file.write('I again,again love Pakistan')
    file.seek(0) # Move cursor to 0 position
    print(file.read())
```

```python
# Append Mode -> Add new content at the end(Creates new file if not exists)
with open('./xyz.txt','a') as file:
    file.write('New file, we are creating')
```

```python
with open('./abc.txt','x') as file:
    file.write('line 1 hjk')

# Only work when file doesn't exist
```

```python
with open('./abc.txt','+rb') as file:
    print(file.read())

# if file doesn't exist it creates new one
```
