```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
churn = pd.read_csv('https://raw.githubusercontent.com/albayraktaroglu/Datasets/master/churn.csv')
churn.head()
```

```python
sns.boxplot(x='Churn?',y='Day Calls',data=churn,color="red")
plt.show()
```

#### Changing order of x-axis

```python
sns.boxplot(x='Churn?',y='Day Calls',data=churn,color="green",order=["True.","False."])
plt.show()
```
