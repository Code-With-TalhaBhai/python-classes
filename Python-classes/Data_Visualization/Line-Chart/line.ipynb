```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.arange(1,11)
x
```

```python
y = x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

### Adding Title And Labels

```python
plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### Changing color Aesthetics

```python
plt.plot(x,y,color='g',linestyle=":",linewidth=4)
plt.show()
```

### Adding two lines in the same plot

```python
x
```

```python
y1=x*2
y2=x*3

plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.plot(x,y2,color='b',linestyle='-',linewidth=3)


plt.title('Two Line Graphs')
plt.grid(True)
plt.show()
```

### Adding Sub-plots

```python
x
```

```python
y1 = x*2
y2 = x*3

# plt.subplot(Total_Rows,Total_Cols,Number_of_subplot)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.grid(True)


plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1)


plt.subplot(2,1,2)
plt.plot(x,y2,color='r')

plt.show()
```
