```python
import matplotlib.pyplot as plt
```

```python
cars = ['BMW','Volvo','Tesla','Lamborghini']

quantity = [54,52,135,18]

plt.pie(quantity,labels=cars)
plt.show()
```

#### Giving specific color to each label

```python
plt.pie(quantity,labels=cars,colors=['red','green','blue','yellow'])
plt.show()
```

#### Show Percentage

```python
plt.pie(quantity,labels=cars,autopct="%0.1f%%")
plt.show()
```

```python
# Upto 4 decimal places
plt.pie(quantity,labels=cars,autopct="%0.4f%%")
plt.show()
```

### Create Doughnut-Chart with Pie-Chart

```python
plt.pie(quantity,labels=cars,radius=2)
plt.pie([23],colors='w',radius=1)
plt.show()
```
