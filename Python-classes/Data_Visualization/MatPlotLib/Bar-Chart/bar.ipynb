```python
import matplotlib.pyplot as plt
```

```python
x = ['a','b']
y = [10,20]

plt.bar(x,y)
plt.show()
```

```python
import numpy as np
x = np.arange(1,11)
y = x * 7

plt.bar(x,y)
plt.show()
```

```python
students = {"Talha":87,"Jazib":64,"Hassan":45}

x = list(students.keys())
y = list(students.values())


plt.bar(x,y,color='g')
plt.show()
```

```python
plt.title('Student Marks')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.bar(x,y,color='y')
plt.show()
```

### Horizontal Bar Charts

```python
plt.barh(x,y,color='r')
plt.show()
```
