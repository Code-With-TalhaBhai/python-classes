```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
one = np.random.randint(0,10,10)
two = np.random.randint(0,10,10)
three = np.random.randint(0,10,10)

data = list([one,two,three])
data
```

```python
plt.boxplot(data)
plt.show()
```
