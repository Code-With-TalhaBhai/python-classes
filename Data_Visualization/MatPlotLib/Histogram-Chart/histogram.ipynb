### Histogram:

* A histogram is used to represent the distribution of a continuous variable.
* In a histogram, the x-axis represents the range of values, and the y-axis represents the frequency or count of observations within each range (bin).
* The bars in a histogram are usually adjacent to each other, with no gaps, to show the continuity of the data.
* Histograms are used to visualize the shape and spread of the data distribution.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
data=np.random.randint(0,500,10) # Generate 10 random numbers between 0-500
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=15)
plt.show()
```
