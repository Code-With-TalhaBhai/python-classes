### Changing Mark Aesthics

```python
import numpy as np

x = np.arange(1,11)
y = x*2

plt.scatter(x,y,marker="*",color='g',s=100) #s->size
plt.show()
```

#### Multiple Plots in Scatter Plots

```python
y1=np.random.randint(0,500,10) # Generate 10 random numbers between 0-500
y2=np.random.randint(0,700,10)

plt.scatter(x,y1,marker="*",color="r",s=70)
plt.scatter(x,y2,marker="^",color="y",s=40)
plt.show()
```

### Subplots in Scatter Plots

```python
plt.subplot(1,2,1)
plt.scatter(x,y1,marker="*",color="r",s=70)
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(x,y2,marker="^",color="y",s=40)
plt.grid(True)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y1,marker="*",color="r",s=70)
plt.grid(True)

plt.subplot(2,1,2)
plt.scatter(x,y2,marker="^",color="y",s=40)
plt.grid(True)

plt.show()
```
