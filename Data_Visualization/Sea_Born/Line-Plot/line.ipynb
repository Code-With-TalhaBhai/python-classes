```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x="signal",y="timepoint",data=fmri)
plt.show()
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
# Check rows and cols in data_set
fmri.shape
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, markers=True, hue="event", style="event")
plt.show()
```
